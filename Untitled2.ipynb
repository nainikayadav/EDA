```python
import matplotlib.pyplot as plt

# Count the number of titles released per year
year_counts = data['release_year'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(12, 6))
year_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Netflix Titles Released by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count of Titles', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the distribution of ratings
rating_counts = data['rating'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
rating_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightblue', 'green'])
plt.title('Distribution of Netflix Titles by Rating', fontsize=16)
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# Count the number of titles per genre
genre_counts = data['listed_in'].str.split(',').explode().value_counts()

# Plot the top 10 genres
plt.figure(figsize=(12, 6))
genre_counts.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Genres on Netflix', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/netflix_titles.csv.zip"
data = pd.read_csv(file_path)

# Check for missing values and unique values in the 'rating' column
print(data['rating'].unique())  # List unique ratings
print(data['rating'].isnull().sum())  # Check for missing values in the 'rating' column

# Clean the data by removing rows with missing or invalid ratings
data_clean = data.dropna(subset=['rating'])

# Count the frequency of each rating
rating_counts = data_clean['rating'].value_counts().sort_values(ascending=False)

# Plotting the Bar Chart for rating distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='viridis')

# Add titles and labels
plt.title('Distribution of Netflix Titles by Rating', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Show the plot
plt.show()
```
